```python
#Questao 1 -> Crie  um  dataframecontendo  apenas  o  seu  tipo  de  imóveis.  Ajuste  os  rótulo  de linhas do dataframe.

import pandas as pd

imovel = pd.read_csv("https://raw.githubusercontent.com/JeffersonCAMarinho/AtividadesIA/main/aluguel.csv", sep= ';')
df = pd.DataFrame(imovel, columns=[ 'Tipo', "Bairro", "Quartos", "Vagas", "Suites", "Area", "Valor", "Condominio", "IPTU"])
boxgaragem = df.loc[imovel["Tipo"] == "Box/Garagem"]

print(boxgaragem)
```

```python
#Questao 2 -> Para  o  novo  dataframe,  proponha  5  possíveis  seleções.  Para  cada  seleção proposta, 
#deixe  um  comentário  no  código  explicando  que  o  objetivo  da  seleção. Exemplo: Selecionar pacientes cujo diagnóstico é doente

#selecionar Box/Garagem no centro

Centro = boxgaragem['Bairro'] == 'Centro'
Centro = boxgaragem[Centro]
Centro

#Selecionar Box/Garagem que possuem vagas
Vagas = boxgaragem['Vagas'] != 0
Vagas = boxgaragem[Vagas]
Vagas

#Selecionar Box/Garagem em que o valor seja menor que R$ 200 
Aluguel = boxgaragem['Valor'] < 200
Aluguel = boxgaragem[Aluguel]
Aluguel

#Selecionar Box/Garagem onde valor do Condominio seja menor que R$ 60
Condominio = boxgaragem['Condominio'] <= 60
Condominio = boxgaragem[Condominio]
Condominio

#Selecionar Box/Garagem onde valor acumulado do condominio + IPTU + valor seja menor que R$ 1000

Total = ((boxgaragem["Valor"] + boxgaragem["Condominio"] + boxgaragem["IPTU"]) <= 1000)
Total = boxgaragem[Total]
Total
```

```python
#Questao 3 -> Chame  o  método  info()  para  o  novo  dataframe,  para  que  seja  possível  verificar quais  atributos  
#possuem  dados  faltantes.  Para  cada  atributo  com  dados  faltante gere um dataframe para exibi-los (utilize o método isnull()).

boxgaragem.info()

boxgaragem.loc[pd.isnull(boxgaragem['Condominio'])]
boxgaragem.loc[pd.isnull(boxgaragem['IPTU'])]
```

```python
#Questao 4 - > Para cada dado faltante analise se 
#eles podem ser removidos ou preenchidos com algum valor. Trate os dados faltantes e justifique o tratamento feito. 
# Resp. sim, pelo fato de se tratar de precos e como eles nao possuem valores, e possivel substituir pelo valor 0

BGsemNulos = boxgaragem.fillna(0)
print(BGsemNulos)
BGsemNulos
```

```python
#Questao 5 -> Proponha  2  possíveis  transformações  para  o  dataframe  utilizando  os  métodos apply() e/ou map()

# Valor do Aluguel pela metade do preco 
BGsemNulos['Valor'] = BGsemNulos['Valor'].map(lambda x: x / 2)
#BGsemNulos

# Todos os Locais onde o Bairro for Centro, Sera Escrito Mosela(Bairro onde morava na Cidade de Petropolis)

BGsemNulos['Bairro'] = BGsemNulos['Bairro'].map({'Centro': 'Mosela'}).fillna(BGsemNulos['Bairro'])
BGsemNulos
```

```python
#Questao 6 -> Salve o dataframe final na extensão csv,
#contendo apenas o seu tipo de imóveis, os tratamentos dos dados faltantes e as transformações feitas.

BGsemNulos.to_csv (r'dataframe.csv', index = None, header=True)
#O CSV fica salvo na guia de arquivos
```
